```python
# Question 32: Find the magnitude of each complex number in list
print("\nQuestion 32: Find the magnitude of each complex number in list")
# Your code here

import cmath
complex_numbers = [3 + 4j, 1 - 1j, -2 + 2j, -3 - 3j]
for num in complex_numbers:
    print(cmath.sqrt(num.real**2 + num.imag**2))
```

```python
# Question 33: Create a nested list with different levels: [1, [2, 3], [4, [5, 6]], 7]
print("\nQuestion 33: Create a nested list with different levels: [1, [2, 3], [4, [5, 6]], 7]")
# Your code here
nested_list = [1, [2, 3], [4, [5, 6]], 7]
print(nested_list)
```

```python
nested_list = [1, [2, 3], [4, [5, 6]], 7]

level = [nested_list]
depth = 0

while len(level) != 0:
    next_level = []

    for item in level:
        if type(item) == list:
            depth = depth + 1
            for x in item:
                if type(x) == list:
                    next_level.append(x)

    level = next_level

print("Depth of nesting:", depth)
```
