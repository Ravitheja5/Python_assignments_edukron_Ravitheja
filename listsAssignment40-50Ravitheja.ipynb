```python
# Question 42: Replace all None values with 0 in list
print("\nQuestion 42: Replace all None values with 0 in list")
# Your code here
result = [None, None, None, None]
l = len(result)
for i in range(l):
    if result[i]==None:
        result[i]=0
print(result)  # Output: [0, 0, 0, 0]
```

```python
# Question 46: Convert each range to list
print("\nQuestion 46: Convert each range to list")
# Your code here
result = [range(3), range(5), range(2)]
empty_list = []
for i in result:
    empty_list.append(list(i))
print(empty_list)  # Output: [[0, 1, 2], [0, 1, 2, 3, 4], [0, 1]]
```
