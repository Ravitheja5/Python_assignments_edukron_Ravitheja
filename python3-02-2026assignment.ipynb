```python
#Count how many vowels are in a given string.


string = input('Enter String To Count Vowels In: ')
vowels = "aeiouAEIOU"
count =0

for i in string:
    if i in vowels:
        count+=1
print(count)
```

```python
#5. Find the factorial of a number.

num = int(input("Enter a Number to find Factorial Of the Number "))

factorial = 1 

for i in range(1,num+1):
    factorial=factorial*i
print(factorial)
```

```python
#6. Check whether a number is prime using a for loop.

num = int(input("Enter a number to check if it is prime: "))


factors =0

for i in range(1,num+1):
    if(num%i==0):
        factors+=1
if factors==2:
    print(f'GIVEN NUMBER IS PRIME NUMBER :{num}')
else:
    print(f'GIVEN NUMBER IS NOT PRIME NUMBER :{num}')
```

```python
#8. Count how many digits are in a number.

num = int(input("Enter a number: "))

string = str(num)
count=0
for i in string:
    if i.isdigit():
        count += 1
print(f"The number of digits in {num} is {count}")  
```

```python
#9. Find the largest number in a list.
#Example: [4, 7, 1, 9, 3]


l = [4, 7, 1, 9, 3]
largest = l[0]
for num in l:
    if num > largest:
        largest = num
print(f"The largest number in the list is {largest}")
```

```python
'''13. Check whether a string is a palindrome.
Example: "madam"'''

string = input("enter string to reverse: ")
l = len(string)
rString = ''

for i in range (l,0,-1):
    rString+=string[i-1]

print(rString)
if rString == string:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
#14. Find duplicate values in a list.
#Example: [1,2,3,2,4,5,1]

Example = [1,2,3,2,4,5,1]


t = tuple(Example)

for i in t:
    if Example.count(i)>1:
        print(f'{i} Is Repeated {Example.count(i)} Times')
```

```python
#15. Print the frequency of each character in a string.
#Example: "banana"'''

string = "banana"

l = list(string)

t = set(l)

for i in t :
    c =l.count(i)
    print(f"Character '{i}' occurs {c} times")
```
